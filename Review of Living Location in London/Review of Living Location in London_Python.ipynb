It is difficult to have clear definition for a good living location as it can vary among individuals depending their financial capability, age, marital status, cultural background, etc. So, let's review the location in London with some general considerations, safety, finance and well-being, and you can make it as reference for getting a good starting point of your new journey.</p>

## 5. Data Analysis

### Rent

```python
#Import packages
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Line plot to show the trend of rent in London
plt.figure(figsize=(10, 6))
sns.set(style = "darkgrid")
sns.lineplot(x = "Year", y = "Rent", data = rent2, hue = "Area")
plt.legend(bbox_to_anchor = (1.05, 1), loc = "upper left")
plt.xticks(rotation = 90)
plt.title("A Line Chart of Rent in London")
plt.show()
```

<br></br>
<p>From 1996 to 2019, the rent of all places in London was increasing.</p>
<br></br>

```python
#Create a table to show top 10 largest rental percentage change areas in London

#Filter
rent3 = rent2.loc[(rent2.Year == "1996-97")]
#Rename of Column
rent3.columns.values[4] = "Rent_1996"
#Select columns
rent3 = rent3.iloc[0:33,[0,1,2,4]]

#Filter
rent4 = rent2.loc[(rent2.Year == "2018-19")]
#Rename of Column
rent4.columns.values[4] = "Rent_2018"
#Select columns
rent4 = rent4.iloc[0:33,[2,4]]

#Join tables
rent4 = pd.merge(rent3, rent4, on = "Area")

#Add percentage change
rent4["Percentage_Change"] = (rent4.Rent_2018-rent4.Rent_1996)/rent4.Rent_1996*100
rent4.sort_values("Percentage_Change", ascending = False).head(10)
```

<br></br>
<p>The largest rent percentage change is in Westminster and increases more than 171%. The top 10 rent percentage changes are all over +100%.</p>
<br></br>

```python
#Show top 10 highest rental areas in London from 1996-2019
rent2.columns.values[4] = "Rent"
rent2.sort_values("Rent", ascending = False).head(10)
```

<br></br>
<p>The highest rental area is Westiminster in 2015-16 that rent is 141.5GBP per week per person. Newham is the most frequent borough in the top 10 list.</p>
<br></br>

```python
#Show top 10 highest rental areas in London in the most recent 5 years(2014-2019)
#Mulitple plots on the same page
sns.set(style = "darkgrid")
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 5))
f, (ax4, ax5) = plt.subplots(1, 2, figsize=(20, 5))

rent5 = rent2.loc[rent2.Year == "2014-15"].sort_values("Rent", ascending = False).head(10)
rent6 = rent2.loc[rent2.Year == "2015-16"].sort_values("Rent", ascending = False).head(10)
rent7 = rent2.loc[rent2.Year == "2016-17"].sort_values("Rent", ascending = False).head(10)
rent8 = rent2.loc[rent2.Year == "2017-18"].sort_values("Rent", ascending = False).head(10)
rent9 = rent2.loc[rent2.Year == "2018-19"].sort_values("Rent", ascending = False).head(10)

sns.barplot(x = rent5.Area, y = rent5.Rent, ax = ax1, color = "#16A085").set(title = "Bar Chart of Top 10 Highest Rental Areas in London in 2014-15", xlabel = "Count", ylabel = "Area") 
sns.barplot(x = rent6.Area, y = rent6.Rent, ax = ax2, color = "#16A085").set(title = "Bar Chart of Top 10 Highest Rental Areas in London in 2015-16", xlabel = "Count", ylabel = "Area") 
sns.barplot(x = rent7.Area, y = rent7.Rent, ax = ax3, color = "#16A085").set(title = "Bar Chart of Top 10 Highest Rental Areas in London in 2016-17", xlabel = "Count", ylabel = "Area")
sns.barplot(x = rent8.Area, y = rent8.Rent, ax = ax4, color = "#16A085").set(title = "Bar Chart of Top 10 Highest Rental Areas in London in 2017-18", xlabel = "Count", ylabel = "Area") 
sns.barplot(x = rent9.Area, y = rent9.Rent, ax = ax5, color = "#16A085").set(title = "Bar Chart of Top 10 Highest Rental Areas in London in 2018-19", xlabel = "Count", ylabel = "Area") 
ax1.set_xticklabels(rent5.Area,Rotation = 90)
ax2.set_xticklabels(rent6.Area,Rotation = 90)
ax3.set_xticklabels(rent7.Area,Rotation = 90)
ax4.set_xticklabels(rent8.Area,Rotation = 90)
ax5.set_xticklabels(rent9.Area,Rotation = 90)
```

<br></br>
<p>Camden, Harrow, Kingston upon Thames, Westminster, Wandsworth and Barnet are in the top 10 highest renta area in London from 2014-2019. Ealing always is in the top 10 list starting from 2015.</p>
<br></br>

```python
#Show top 10 lowest rental areas in London in the most recent 5 years(2014-2019)
#Mulitple plots on the same page
sns.set(style = "darkgrid")
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 5))
f, (ax4, ax5) = plt.subplots(1, 2, figsize=(20, 5))

rent5 = rent2.loc[rent2.Year == "2014-15"].sort_values("Rent", ascending = True).head(10)
rent6 = rent2.loc[rent2.Year == "2015-16"].sort_values("Rent", ascending = True).head(10)
rent7 = rent2.loc[rent2.Year == "2016-17"].sort_values("Rent", ascending = True).head(10)
rent8 = rent2.loc[rent2.Year == "2017-18"].sort_values("Rent", ascending = True).head(10)
rent9 = rent2.loc[rent2.Year == "2018-19"].sort_values("Rent", ascending = True).head(10)

sns.barplot(x = rent5.Area, y = rent5.Rent, ax = ax1, color = "#16A085").set(title = "Bar Chart of Top 10 Lowest Rental Areas in London in 2014-15", xlabel = "Count", ylabel = "Area") 
sns.barplot(x = rent6.Area, y = rent6.Rent, ax = ax2, color = "#16A085").set(title = "Bar Chart of Top 10 Lowest Rental Areas in London in 2015-16", xlabel = "Count", ylabel = "Area") 
sns.barplot(x = rent7.Area, y = rent7.Rent, ax = ax3, color = "#16A085").set(title = "Bar Chart of Top 10 Lowest Rental Areas in London in 2016-17", xlabel = "Count", ylabel = "Area")
sns.barplot(x = rent8.Area, y = rent8.Rent, ax = ax4, color = "#16A085").set(title = "Bar Chart of Top 10 Lowest Rental Areas in London in 2017-18", xlabel = "Count", ylabel = "Area") 
sns.barplot(x = rent9.Area, y = rent9.Rent, ax = ax5, color = "#16A085").set(title = "Bar Chart of Top 10 Lowest Rental Areas in London in 2018-19", xlabel = "Count", ylabel = "Area") 
ax1.set_xticklabels(rent5.Area,Rotation = 90)
ax2.set_xticklabels(rent6.Area,Rotation = 90)
ax3.set_xticklabels(rent7.Area,Rotation = 90)
ax4.set_xticklabels(rent8.Area,Rotation = 90)
ax5.set_xticklabels(rent9.Area,Rotation = 90)
```
