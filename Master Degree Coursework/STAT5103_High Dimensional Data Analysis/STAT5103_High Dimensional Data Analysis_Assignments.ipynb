```r
#Question 1a
#Data Import
load("decathlon.RData")
str(decathlon)


#Question 1b
#Calculation of variance ratio (max/min)
(q1b<-apply(decathlon[,1:10],2,var))
(min_q1b<-min(q1b))
(max_q1b<-max(q1b))
(ratio_q1b<-max_q1b/min_q1b)


#Question 1d
#PCA
(PCA<-prcomp(decathlon[,1:10],scale. = TRUE))

#Question 1e
#Display reduced data
head(PCA$x,3)

#Question 1f
#Display cumulative proportion of variance
summary(PCA)

#Question 1g
#Display relationship between each PC and variable
PCA$rotation

#Question  1h
#Scatter plot
plot(PCA$x[,1],decathlon$Points,col=as.numeric(decathlon$Competition),xlab="First PC",ylab="Point")
legend("bottomright",c("2004 Decastar","2004 Olympic Game"),pch=1,col=c(1,2),cex=0.8)
```

<br></br>
<p>In answer (b), found that the ratio is large because 10 events in “decathlon” have different units, for example, long jump measured by length, 100m race measured by time, and the variance difference is large. PC technique is sensitive to scale change because the eigenvalue decomposition is performed on the variance matrix. Component variables of a random vector X are general heterogeneous with respect to their variances. Therefore, components of X are not equally weighted. Variable with larger variance has larger effect on PC. To remedy the problem, standardization is recommended before PCA.</p>
<br></br>
<p>From R’s result, 7 principal components (PC1 to PC7) are needed to keep in order to explain at least 90% of the total variation. It is not surprising as the design of the decathlon is to include events that require diverse skills and strength.</p>
<br></br>
<p>Judging from the sign of the coefficients for the first principal component, it is the difference in the performance in field events (having positive coefficients) and in track events (having negative coefficients).</p>

```r
#Question 3a
#Compression of photo
install.packages("imager")
library(imager)

#Load photo
photo<-load.image("rabbit.jpg")
str(photo)
```

<br>
<p>Object photo is a d-dimensional array. The pixel values for each colour (red, green blue) are stored in an m×n matrix where m=1024 nad n=576. If p principal components are selected, the compressed ratio is [n+p(m+n)] ⁄ (mn). If the ratio is less than 0.1, p=36.</p>
</br>

```r
#PCA for each color
r<-photo[,,1,1]
g<-photo[,,1,2]
b<-photo[,,1,3]

photo.r<-prcomp(r)
photo.g<-prcomp(g)
photo.b<-prcomp(b)

#Number of PCs used
k<-36

#Store center x[,1:k] and rotation [,1,k] for each colour
comp.r<-matrix(photo.r$center,nrow = dim(r)[1],ncol = dim(r)[2], byrow = TRUE) + photo.r$x[,1:k]%*%t(photo.r$rotation[,1:k])
comp.g<-matrix(photo.g$center,nrow = dim(g)[1],ncol = dim(g)[2], byrow = TRUE) + photo.g$x[,1:k]%*%t(photo.g$rotation[,1:k])
comp.b<-matrix(photo.b$center,nrow = dim(b)[1],ncol = dim(b)[2], byrow = TRUE) + photo.b$x[,1:k]%*%t(photo.b$rotation[,1:k])

#Create cimg
Cphoto<-as.cimg(rep(0,prod(dim(photo))),dim = dim(photo))
Cphoto[,,1,1]<-comp.r
Cphoto[,,1,2]<-comp.g
Cphoto[,,1,3]<-comp.b

#Comparison for 2 photos
par(mfrow = c(1,2))
plot(photo,main = "Original Photo")
plot(Cphoto, main = paste("Use ",k,"PC"))
par(mfrow = c(1,1))
save.image(Cphoto,"new_rabbit.jpeg")

#Compression Ration
summary(photo.r)$importance[,k]
summary(photo.g)$importance[,k]
summary(photo.b)$importance[,k]

compression.ratio<-3*(length(photo.r$center)+prod(dim(photo.r$x[,1:k]))+prod(dim(photo.r$rotation[,1:k])))/prod(dim(photo))
compression.ratio
```

```r
#Question 2
#Data Import
load("decathlon.RData")
str(decathlon)


#Question 2a
#Perform factor analysis to find maximum number of factors
for (i in 1:10) {
  q2a<-factanal(decathlon[,1:10],i,scores = "regression")
  print(q2a)
}
```

<p>The first factor measures the difference in performance in track events (taking positive sign) and field events (taking negative sign).</p>
