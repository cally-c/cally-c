```python
#Question 2
#US Census Aggregated Dataset

#Import packages

import pandas as pd
import numpy as np

#Read file
census_df = pd.read_csv("census.csv")
census_df.head()
```

```python
#Question A: Which state has the most counties in it?
#Filter sumlev=50
#Count no. of sumlev by state names
#Return the max. value with state name and no. of sumlev
def questionA():
    a=census_df[census_df['SUMLEV']==50]
    a=a.groupby('STNAME').count()['SUMLEV']
    return a[a==a.max()]
questionA()
```

```python
#Question B: The three most populous states (which count the population in all of the constitute counties)
#COUNTY == 0 means the whole state in this dataset.
#Filter sumlev=50
#Sum census2010pop by state names
#Return top three 
def questionB():
    b=census_df[census_df['SUMLEV']==50]
    b=b.groupby('STNAME').sum()['CENSUS2010POP'].nlargest(n=3)
    return b
questionB()
```

```python
#Question C: Which county has had the largest absolute change in population within the period 2010-2015?
#Filter sumlev=50
#Find the max. and min. values in population within the period 2010-2015 & calculate the largest absolute change
#Find the largest absolute change of county
def questionC():
    c=census_df[census_df['SUMLEV']==50]
    c=c[['CTYNAME','POPESTIMATE2010', 'POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']]
    max_c=map(max,c['POPESTIMATE2010'],c['POPESTIMATE2011'],c['POPESTIMATE2012'],c['POPESTIMATE2013'],c['POPESTIMATE2014'],c['POPESTIMATE2015'])
    max_c=np.array(list(max_c))
    min_c=map(min,c['POPESTIMATE2010'],c['POPESTIMATE2011'],c['POPESTIMATE2012'],c['POPESTIMATE2013'],c['POPESTIMATE2014'],c['POPESTIMATE2015'])
    min_c=np.array(list(min_c))
    absolute_change=max_c-min_c
    return c.iloc[absolute_change.argmax()].values[0]
questionC()
```

```python
#Question D: Create a query that finds the counties that belong to regions 1 or 2, 
#whose name starts with 'Washington', and whose POPESTIMATE2015 was greaterthan their POPESTIMATE2014 .
def questionD():
    d=census_df[(census_df['SUMLEV']==50) & (census_df['REGION']<=2) & (census_df['CTYNAME'].str.match('^Washington')) & (census_df['POPESTIMATE2015']>census_df['POPESTIMATE2014'])]
    return d
questionD()
```
